```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
url = "https://raw.githubusercontent.com/samg050213sh/kz-kr-life-journey/main/data/life_data.csv"
df = pd.read_csv(url, encoding='latin1')
df
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['year'], df['korean_level'], marker='o')
plt.title('Korean Level Over Years')
plt.xlabel('Year')
plt.ylabel('Korean Level')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['year'], df['stress'], marker='o', label='Stress')
plt.plot(df['year'], df['happiness'], marker='o', label='Happiness')
plt.title('Stress and Happiness Over Years')
plt.xlabel('Year')
plt.ylabel('Level')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['year'], df['nostalgia_kz'], marker='o', label='Nostalgia for Kazakhstan')
plt.plot(df['year'], df['belonging_kr'], marker='o', label='Belonging to Korea')
plt.title('Emotional Transition Over Years')
plt.xlabel('Year')
plt.ylabel('Level')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['year'], df['friends_count'])
plt.title('Number of Friends Over Years')
plt.xlabel('Year')
plt.ylabel('Friends Count')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['year'], df['sleep_quality'], marker='o')
plt.title('Sleep Quality Over Years')
plt.xlabel('Year')
plt.ylabel('Sleep Quality')
plt.grid(True)
plt.show()
```

```python
numeric_cols = ['year', 'age', 'korean_level', 'stress', 'happiness',
                'nostalgia_kz', 'belonging_kr', 'friends_count', 'sleep_quality']

corr = df[numeric_cols].corr()
corr
```

```python
corr.to_csv("/content/kz-kr-life-journey/results/correlation_matrix.csv", index=True)
```

```python
from sklearn.linear_model import LinearRegression

features = ['stress', 'friends_count', 'sleep_quality', 'korean_level',
            'nostalgia_kz', 'belonging_kr']

X = df[features]
y = df['happiness']

model = LinearRegression()
model.fit(X, y)

print("Coefficients:")
for name, coef in zip(features, model.coef_):
    print(f"{name}: {coef:.3f}")

print("Intercept:", model.intercept_)
```

```python
df['happiness_pred'] = model.predict(X)
df[['year', 'happiness', 'happiness_pred']]
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['year'], df['happiness'], marker='o', label='Real happiness')
plt.plot(df['year'], df['happiness_pred'], marker='o', linestyle='--', label='Predicted happiness')
plt.title('Real vs Predicted Happiness')
plt.xlabel('Year')
plt.ylabel('Happiness Level')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig(save_path + "/real_vs_predicted_happiness.png", dpi=300, bbox_inches='tight')
```
